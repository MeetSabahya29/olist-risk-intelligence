```python
import pandas as pd

# ૧. તમારો મેઈન ડેટાસેટ લોડ કરો
df = pd.read_csv("olist_final_dataset.csv")

# ૨. ડેટાને પૂરેપૂરો Shuffle (મિક્સ) કરો
# random_state=42 રાખવાથી દર વખતે સરખો જ મિક્સ ડેટા મળશે
df_shuffled = df.sample(frac=1, random_state=42).reset_index(drop=True)

# ૩. મિક્સ થયેલા ડેટામાંથી પર્સનલ ૫૦,૦૦૦ રો (Rows) અલગ કરો
df_final_small = df_shuffled.head(35000)

# ૪. તેને ઓરિજિનલ નામથી જ સેવ (Overwrite) કરી દો
df_final_small.to_csv("olist_final-1_dataset.csv", index=False)
```
